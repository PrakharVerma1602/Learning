```python
import requests
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/Human_body_weight'
```

```python
response=requests.get(url,headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'})
```

```python
response
```

```python
response.text
```

```python
from bs4 import BeautifulSoup
```

```python
out=BeautifulSoup(response.text,"html.parser")
```

```python
out
```

```python
l=out.find_all('table',{'class':'wikitable'})
```

```python
str(l[0])
```

```python
a=pd.read_html(str(l[0]))
```

```python
df=a[0]
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['Ref'],inplace=True)
```

```python
df.head()
```

```python
df['Average weight']
```

```python
a='60.7 kg (130.8 lb)'
```

```python
a.split("(")[0]
```

```python
import re
```

```python
float(re.findall("\d+\.\d+",a.split("(")[0])[0])
```

```python
df['Average weight'] = df['Average weight'].apply(lambda x: float(re.findall(r"\d+\.\d+", x.split("(")[0])[0]))
```
